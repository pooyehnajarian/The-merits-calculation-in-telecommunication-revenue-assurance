# **MTTD**

```python
df3= pd.read_excel('/content/RAMSReport.xlsx')
df3.head(5)
```

```python
df31 =  df3.loc[(df3.IssueType == "Change Related")]
df31
```

```python
from datetime import datetime
df31['identifiedDate'] = pd.to_datetime(df31['identifiedDate'])
df31['firstOccurrenceDate'] = pd.to_datetime(df31['firstOccurrenceDate'])
```

```python
df31['MTTD']=df31['identifiedDate']-df31['firstOccurrenceDate']
```

```python
df31
```

```python
#df['days'] = df['duration'].dt.days
df31['MTTDNew']=df31['MTTD'].dt.days
df31
```

```python
TotalLeakage=df31['Local Currency Leakage Detected Value'].sum()
TotalLeakage
```

```python
df31['Rate']=(df31['Local Currency Leakage Detected Value'])/TotalLeakage
df31
```

```python
df31['MTTDRate']= df31['Rate']*df31['MTTDNew']
MTDDCal=df31['MTTDRate'].sum()
MTDDCal
```

```python
datatoexcel = pd.ExcelWriter('MTTD.xlsx')
df31.to_excel(datatoexcel)
datatoexcel.save()
```

# MTTR

```python
df3= pd.read_excel('/content/RAMSReport.xlsx')
df3.head(5)
```

```python
df32 =  df3.loc[(df3.IssueType == "Change Related")]
df32
```

```python
from datetime import datetime
df32['identifiedDate'] = pd.to_datetime(df32['identifiedDate'])
df32['lastOccuranceDate'] = pd.to_datetime(df32['lastOccuranceDate'])
```

```python
df32['MTTR']=df32['lastOccuranceDate']-df32['identifiedDate']
```

```python
df32['MTTRNew']=df32['MTTR'].dt.days
df32
```

```python
TotalLeakage=df32['Local Currency Leakage Detected Value'].sum()
TotalLeakage
```

```python
df32['Rate']=(df32['Local Currency Leakage Detected Value'])/TotalLeakage
df32
```

```python
df32 = df32._get_numeric_data()
df32[df32 < 0] = 0
df32
```

```python
df32['MTTRRate']= df32['Rate']*df32['MTTRNew']
```

```python
MTTRCal=df32['MTTRRate'].sum()
MTTRCal
```

```python
datatoexcel = pd.ExcelWriter('MTTR.xlsx')
df32.to_excel(datatoexcel)
datatoexcel.save()
```
